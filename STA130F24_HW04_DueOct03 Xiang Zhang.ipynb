# Difference：
The standard deviation shows the distribution and change of data points around the mean and shows us how dispersed the data is. If the standard deviation is large, the difference between the data point and the mean is large, and if not, it is small. The standard error of the mean reflects the expected deviation between the mean of the sample after sampling survey and the mean of the real data collection.
In short, standard deviation shows the distance distribution between a single data point and the mean; The standard error of mean shows the approximate distribution of the mean of the sample.
In addition, the calculation formulas of the two are also slightly different. standard error of the mean is divided by the square root of the observed quantity on the basis of the standard deviation formula.

# Distinct ideas：
The standard deviation shows how dense a given data point or collected data is from the mean in the visualization, regardless of the size of the sample. The standard of error of the mean is the accuracy of the estimate of the population mean.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Sample height data
data = {'ID': range(1, 21),
        'Height(cm)': [160, 165, 170, 172, 168, 174, 162, 169, 171, 173, 166, 175, 167, 164, 176, 169, 171, 168, 172, 170]}

df = pd.DataFrame(data)

# Function to take 5 random samples and calculate their sample mean
def random_sample_mean(df, n_samples=5):
    samples = df['Height(cm)'].sample(n=n_samples, replace=False)  # Take random sample
    sample_mean = np.mean(samples)  # Calculate the mean
    return sample_mean

# Example: Running the function multiple times
n_trials = 1000  # Number of times to repeat the sampling
sample_size = 5
sample_means = [random_sample_mean(df, sample_size) for _ in range(n_trials)]

# Calculate standard error of the mean (SEM)
sample_std = np.std(sample_means, ddof=1)  # Sample standard deviation
sem = sample_std / np.sqrt(sample_size)  # Standard error of the mean

print(f"Standard Error of the Mean (SEM): {sem:.2f}")

# Plotting the distribution of sample means
sns.histplot(sample_means, kde=True, stat="density", color="skyblue", label="Sample Mean Distribution")
mean_of_means = np.mean(sample_means)
std_of_means = np.std(sample_means)

# Plot normal distribution for comparison
x_vals = np.linspace(mean_of_means - 4*std_of_means, mean_of_means + 4*std_of_means, 100)
plt.plot(x_vals, norm.pdf(x_vals, mean_of_means, std_of_means), label='Normal Distribution', color='red')

# Add labels and title
plt.title('Distribution of Sample Means with Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

# Sample height data
data = {'ID': range(1, 21),
        'Height(cm)': [160, 165, 170, 172, 168, 174, 162, 169, 171, 173, 166, 175, 167, 164, 176, 169, 171, 168, 172, 170]}

df = pd.DataFrame(data)

# Calculate mean and standard deviation
mean_height = df['Height(cm)'].mean()
std_height = df['Height(cm)'].std(ddof=1)  # Sample standard deviation
n = len(df['Height(cm)'])  # Sample size

# Calculate the standard error of the mean (SEM)
sem = std_height / np.sqrt(n)

# Determine the critical value for 95% confidence interval
critical_value = stats.t.ppf(0.975, df=n-1)  # Two-tailed

# Calculate the margin of error
margin_of_error = critical_value * sem

# Confidence intervals
confidence_interval = (mean_height - margin_of_error, mean_height + margin_of_error)

# Display results
print(f"Mean Height: {mean_height:.2f} cm")
print(f"95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f}) cm")

# Plotting
plt.figure(figsize=(8, 5))
plt.errorbar(x=1, y=mean_height, yerr=margin_of_error, fmt='o', 
             color='blue', label='Mean Height with 95% CI', capsize=5)

plt.xticks([1], ['Average Height'])
plt.ylabel('Height (cm)')
plt.title('Mean Height with 95% Confidence Interval')
plt.axhline(y=confidence_interval[0], color='red', linestyle='--', label='95% CI Lower Bound')
plt.axhline(y=confidence_interval[1], color='green', linestyle='--', label='95% CI Upper Bound')
plt.legend()
plt.grid()
plt.ylim(confidence_interval[0] - 5, confidence_interval[1] + 5)

# Show the plot
plt.show()
````

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# Check the current working directory
print("Current Working Directory:", os.getcwd())

# If your file is not in the current working directory, either move the file there or change the directory:
# os.chdir('/path/to/your/folder')  # Uncomment and modify this if needed

# Load the CSV data file (ensure the file is in the correct directory or provide the full path)
file_path = 'vaccine_data.csv'  # Replace this with your actual file path if needed
try:
    data = pd.read_csv(file_path)  # Attempt to load the file
    print("File loaded successfully!")
except FileNotFoundError:
    print(f"File not found: {file_path}. Please check the file path and name.")
    print("Files in current directory:", os.listdir(os.getcwd()))
    data = None  # Set data to None to prevent further processing if the file is missing

# Proceed only if the data was loaded successfully
if data is not None:
    # Set seed for reproducibility
    np.random.seed(42)

    # Calculate difference between Final and Initial Health Score
    data['Difference'] = data['FinalHealthScore'] - data['InitialHealthScore']

    # Bootstrapping
    bootstrapped_means = []
    for _ in range(1000):
        sample = data['Difference'].sample(frac=1, replace=True)
        bootstrapped_means.append(sample.mean())

    # Calculate 95% Confidence Interval
    ci_lower = np.percentile(bootstrapped_means, 2.5)
    ci_upper = np.percentile(bootstrapped_means, 97.5)

    # Plot the bootstrap distribution
    plt.hist(bootstrapped_means, bins=30, edgecolor='k')
    plt.axvline(ci_lower, color='red', linestyle='--', label=f'Lower CI: {ci_lower:.2f}')
    plt.axvline(ci_upper, color='blue', linestyle='--', label=f'Upper CI: {ci_upper:.2f}')
    plt.title('Bootstrap Distribution of Mean Health Score Differences')
    plt.legend()
    plt.show()

    # Print out the confidence interval
    print(f"95% Confidence Interval for the mean difference: [{ci_lower:.2f}, {ci_upper:.2f}]")
else:
    print("Exiting the program due to file loading error.")
````
